```python
import matplotlib.pyplot as plt
days=[1,2,3,4,5,6,7]
nop1=[5,10,30,20,35,60,80]
nop2=[10,20,30,25,40,50,90]
nop3=[8,30,50,60,70,90,100]
plt.stackplot(days, nop1,nop2,nop3)
plt.show()
```

```python
plt.stackplot(days, nop1,nop2,nop3,colors=["hotpink","black","yellow"])
plt.show()
```

```python
plt.stackplot(days, nop1,nop2,nop3,labels=["week1","week2","week3"])
plt.legend()
plt.show()
```
