```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
data=sns.load_dataset("tips")
print(data)
```

```python
sns.set_style(style="darkgrid")

sns.barplot(data,x="day",y="tip")
plt.show()
```

```python
sns.set_style(style="whitegrid")

sns.barplot(data,x="day",y="tip")
plt.show()
```

```python
sns.barplot(data,x="day",y="tip",hue="sex",palette="spring")
plt.show()
```
