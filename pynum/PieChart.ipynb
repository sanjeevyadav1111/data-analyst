```python
import matplotlib.pyplot as plt
brands=["oneplus","nokia","samsung","redmi","apple"]
popularity=[22,35,30,3,10]
c=["red","blue","green","yellow","purple"]
plt.pie(popularity,labels=brands,colors=c,shadow=True)
plt.show()
```

```python
plt.pie(popularity,labels=brands,colors=c,shadow=True,autopct="%.2f")
plt.show()
```

```python
plt.pie(popularity,labels=brands,colors=c,shadow=True,autopct="%.2f",startangle=90)
plt.show()
```
