```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
data=sns.load_dataset("tips")
print(data)
```

```python
# sns.heatmap(data,vmax=40,vmin=1)
gp=data.groupby("day").agg({"tip":"mean"})
sns.heatmap(gp)
plt.show()
```

```python
sns.pairplot(data)
plt.show()
```

```python
sns.pairplot(data,hue="day")
plt.show()
```
